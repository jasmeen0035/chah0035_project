```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Qus-A

```python
# Image Read and show on Grayscale mode
img = cv2.imread('assets/skull.png', 0)
```

```python
# Taking User Input to Reduce Intensity Levels
print(' Choose Input between 1 to 7\n')
k = int(input('Enter the number for Reduce Intensity Levels : '))
```

```python
# Intensity Level is 2^k

Intensity_Level = 2**k
print('\nUser Input vale is :', k)
print('\nIntensity_Level when 2^k :', Intensity_Level)
```

```python
Intensity_factor = 256 / Intensity_Level

# New Intensity Image

New_Intensity_img = np.uint8(np.floor(np.double(img)/np.max(image) * Intensity_Level) * Intensity_factor)

# Save Image using opencv imwrite
plt.imsave('Output/int_' + str(k) + 'user_.png', New_Intensity_img, cmap='gray')
```

```python
plt.imshow(New_Intensity_img, cmap='gray')
```
